```python
from subprocess import Popen, PIPE

import numpy as np
import matplotlib.pyplot as plt
```

```python
scoress = {}
timess = {}
```

```python
!clang++ -std=c++17 -Wall -Wextra ../answer/main.cpp -O2 -g -I/include -DNDEBUG
```

```python
scores = []
times = []
N_TESTCASES = 500
name = "optuna_004_110_trial3"
assert name not in scoress
for i in range(N_TESTCASES):
    i = f"{i:04d}"
    print(i)
    res = !time -p ../tools/target/release/tester ./a.out < ../tools/in/{i}.txt > out.txt
    
    score = int(res[-4].split()[-1])
    t = float(res[-3].split()[-1])
    
    scores.append(score)
    times.append(t)
    
    print(res)

scoress[name] = scores
timess[name] = times
```

```python
with open(f"scores_{name}.txt", "w") as f:
    print(*scores, file=f)
```

```python
Rs = []
for i in range(N_TESTCASES):
    i = f"{i:04d}"
    a = !head ../tools/in/{i}.txt -n 1
    R = int(a[0].split()[3])
    Rs.append(R)
```

```python
plt.figure(figsize=(8, 6))
for nm, scrs in scoress.items():
    plt.scatter(Rs, scrs, s=3)
    xs = np.arange(1000, 3000, 20)
    ys = np.poly1d(np.polyfit(Rs, scrs, 2))(xs)
    plt.plot(xs, ys, label=nm)
plt.grid()
plt.legend()
plt.show()
```
